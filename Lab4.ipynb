**k means clustering algorithm **

Generating 10,000 2-D data points randomly.

```python
import numpy as np
import time
from sklearn.cluster import KMeans
data = np.random.rand(10000, 2)
kmeans = KMeans(n_clusters=5)
```

inding the time taken by the algorithm

```python
start_time = time.time()
kmeans.fit(data)
end_time = time.time()
time_taken = end_time - start_time
print(f"Time taken by K-means to find clusters: {time_taken:.4f} seconds")
```

finding the cluster center

```python
print("Cluster Centers:\n", kmeans.cluster_centers_)
```

**Implementinig mini batch k-means clustering algorithm**

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Generating 10,000 2-D data points in the range 0-100 randomly and into 5  clusters.



```python
data = np.random.rand(1000, 2)
kmeans = KMeans(n_clusters=5)
```

fit the model


```python
kmeans.fit(data)
```

plot the data point and cluster center

```python
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.title("K-means Clustering")
plt.show()
```

**Finding Clusters of Iris Dataset using KMedoids Algorithm**

```python
import matplotlib.pyplot as plt
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.utils import read_sample
from sklearn.datasets import load_iris
import numpy as np
```

loading data

```python
iris = load_iris()
data = iris.data
```

select initial mediods

```python
initial_medoids = [0, 50, 100]
```

create k mediods instance


```python
kmedoids_instance = kmedoids(data, initial_medoids)
```

running clustering and getting results

```python
kmedoids_instance.process()
clusters = kmedoids_instance.get_clusters()
labels = np.zeros(len(data))
```

assigning cluster labels

```python
for cluster_id, cluster in enumerate(clusters):
    for index in cluster:
        labels[index] = cluster_id
```

plotting clusters

```python
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title("KMedoids Clustering (pyclustering) on Iris Dataset")
plt.show()
```
